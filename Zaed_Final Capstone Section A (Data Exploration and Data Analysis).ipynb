iii. Display the average of these attributes: 'Age', 'Monthly Income', and 'Years at Company'.

```python
# Calculate the average

avg_norm_loss = data["Age"].astype("float").mean()
print("Average of Age: {:.2f}".format(avg_norm_loss))

avg_norm_loss = data["MonthlyIncome"].astype("float").mean()
print("Average of Monthly Income: {:.2f}".format(avg_norm_loss))

avg_norm_loss = data["YearsAtCompany"].astype("float").mean()
print("Average of Years at Company: {:.2f}".format(avg_norm_loss))
```

iv. Find the minimum and maximum 'Monthly Income'

```python
# Find the minimum and maximum 'Monthly Income'

min_monthly_income = data['MonthlyIncome'].min()
max_monthly_income = data['MonthlyIncome'].max()

print("Minimum Monthly Income:", min_monthly_income)
print("Maximum Monthly Income:", max_monthly_income)
```

v. Provide graphical visualization by plotting Histogram of 'Monthly Income vs. staff numbers'

```python
# Plot histogram

plt.figure(figsize=(10, 6))
plt.hist(data['MonthlyIncome'], bins=20, color='green', edgecolor='black', alpha=0.7)
plt.xlabel('Monthly Income')
plt.ylabel('Number of Employees')
plt.title('Monthly Income vs. Number of Employees')
plt.grid(True)
plt.show()
```

vi. Provide graphical Visualization of the distribution between 'Year at Company' and 'Monthly Income' using the scatter plot

```python
# Plot scatter plot

plt.figure(figsize=(10, 6))
plt.scatter(data['YearsAtCompany'], data['MonthlyIncome'], color='orange', alpha=0.7)
plt.xlabel('Years at Company')
plt.ylabel('Monthly Income')
plt.title('Distribution between Years at Company and Monthly Income')
plt.grid(True)
plt.show()
```

vii. Find the correlation between 'Years at Company' and 'Monthly Income'

```python
# Find the correlation between 'Years at Company' and 'Monthly Income'

correlation = data['YearsAtCompany'].corr(data['MonthlyIncome'])
print("Correlation between Years at Company and Monthly Income:", correlation)
```
